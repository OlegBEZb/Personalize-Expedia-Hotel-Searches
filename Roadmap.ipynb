```python
public_scores = [0.32661, 0.36825, 0.36995, 0.36163, 0.33767, 0.31550, 0.33190, 0.38823]
local_test_scores = [0.32595, 0.78699, 0.37916, 0.36886, 0.37758, 0.35682, 0.38264, 0.38736]
local_val_scores = [np.nan, np.nan, np.nan, np.nan, 0.36971, 0.35554, 0.38184, 0.38674]
local_train_scores = [np.nan, np.nan, np.nan, np.nan, 0.12225, 0.07871, 0.18755, 0.43494]
```

# All

```python
def plot(public, test, val, train):
    plt.figure(figsize=(10, 5))
    plt.plot(public, marker='+', label='public')
    plt.plot(test, marker='+', label='local test')
    plt.plot(val, marker='+', label='local val')
    plt.plot(train, marker='+', label='local train')
    plt.xlabel('submission', fontsize=14)
    plt.xticks(np.arange(len(public)), np.arange(1, len(public)+1))
    plt.ylabel('NDCG', fontsize=14)
    plt.title('NDCG scores')
    plt.legend(fontsize=12);
    
plot(public=public_scores, 
     test=local_test_scores, 
     val=local_val_scores, 
     train=local_train_scores)
```
